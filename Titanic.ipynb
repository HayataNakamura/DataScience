```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, Imputer
import math
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('train.csv')
df['Embarked'].value_counts()
```

```python
df1 = pd.read_csv('test.csv')
```

```python
emb_mapping = {'S': 1, 'C': 2, 'Q': 3}
df['Embarked'] = df['Embarked'].map(emb_mapping)
df['Embarked'] = df['Embarked'].fillna(1)

df1['Embarked'] = df1['Embarked'].map(emb_mapping)
df1['Embarked'] = df1['Embarked'].fillna(1)
```

```python
df['Age'].values
```

```python
imr = Imputer(missing_values = 'NaN', strategy = 'mean')
imr = imr.fit(df[['Age']])
imputed_data = imr.transform(df[['Age']])
df['Age']=imputed_data

imr1 = imr.fit(df1[['Age']])
imputed_data1 = imr1.transform(df1[['Age']])
df1['Age']=imputed_data1

imr2 = imr.fit(df1[['Fare']])
imputed_data2 = imr2.transform(df1[['Fare']])
df1['Fare']= imputed_data2
```

```python
sex_mapping = {'male': 1, 'female': 0}
df['Sex'] = df['Sex'].map(sex_mapping)
df1['Sex'] = df1['Sex'].map(sex_mapping)
```

```python
for i, name in enumerate(df['Name'].values):
    df['Name'][i] = len(name)
df['Name']
```

```python
for i, name in enumerate(df1['Name'].values):
    df1['Name'][i] = len(name)
```

```python
df['Cabin'] = df['Cabin'].fillna(0)
df1['Cabin'] = df1['Cabin'].fillna(0)

for i in range(len(df['Cabin'])):
    if df['Cabin'][i] != 0:
        df['Cabin'][i] = 1
for i in range(len(df1['Cabin'])):
    if df1['Cabin'][i] != 0:
        df1['Cabin'][i] = 1
```

```python
df = df.drop('Ticket', axis=1)
df1 = df1.drop('Ticket', axis=1)
```

```python
X_train, y_train = df.iloc[:, 2:], df.iloc[:, 1]
X_test = df1.iloc[:, 1:]
```

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
lr = LogisticRegression(C=100.0, random_state=1)
lr.fit(X_train_std, y_train)
```

```python
predictions = lr.predict(X_test_std)
```
